```python
import pandas as pd
import numpy as np
import yfinance as yf
import math
```

```python
tickers=pd.read_csv(r"C:\Users\KIIT\Desktop\Code\top_50_indian_stocks.csv")
tickers.head()
```

```python
def fetch_market_cap(ticker_list):
    data = yf.download(ticker_list, period = '1d', group_by='ticker', auto_adjust=False)
    
    stocks_data = []
    for ticker in ticker_list:
        latest_price = data[ticker]['Close'].iloc[-1]
        market_cap = yf.Ticker(ticker).info.get("marketCap", 'N/A') #if not found then print N/A
        stocks_data.append({    #key value pair so that we can convert it into dataframe later easily
            "Ticker" : ticker,
            "Market Cap" : market_cap,
            "Latest Price" : latest_price,
        })
        
    stocks_df = pd.DataFrame(stocks_data)
        
    return stocks_df
        
    

tickers_list = tickers["Ticker"].values.tolist()
df = fetch_market_cap(tickers_list)
df
```

```python
#SORT DATA ACC TO THEIR MARKET CAP

df=df.sort_values(by='Market Cap', ascending=False)
df
```
